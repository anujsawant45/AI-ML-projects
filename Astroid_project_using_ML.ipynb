```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv('neo.csv')
```

```python
a
```

```python
a.drop(['id','name'],axis=1,inplace=True)
```

```python
a
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit(a['orbiting_body'])
a['orbiting_body']=le.transform(a['orbiting_body'])
a['orbiting_body']=a['orbiting_body'].astype(float)
```

```python
le.fit(a['sentry_object'])
a['sentry_object']=le.transform(a['sentry_object'])
a['sentry_object']=a['sentry_object'].astype(float)
```

```python
le.fit(a['hazardous'])
a['hazardous']=le.transform(a['hazardous'])
a['hazardous']=a['hazardous'].astype(float)
```

```python
a.dtypes
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(a.corr(),annot=True,cbar=False)
plt.show()
```

```python
x=a.iloc[:,:-1]
y=a.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

# Linear Regression


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
lr=LinearRegression()
```

```python

lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
residual=y_test-y_pred
plt.figure(figsize=(5,5))
plt.scatter(y_pred,residual)
plt.show()
```

```python
plt.figure
sns.distplot(residual)
plt.plot()
```

```python
lr.score(x_test,y_test)
```

```python
lr.coef_
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2_score=r2_score(y_test,y_pred)
```

```python
print(mae)
print(mse)
print(rmse)
print(r2_score)
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assume x and y are your features and labels
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Feature scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Optional: Dimensionality reduction
# pca = PCA(n_components=0.95)  # Keep 95% of variance
# x_train = pca.fit_transform(x_train)
# x_test = pca.transform(x_test)

# Train the SVM
clfs = svm.SVC(kernel='linear', gamma='auto', C=2)
clfs.fit(x_train, y_train)
```

```python
y_preds=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
score=accuracy_score(y_preds,y_test)
print(score)
```
